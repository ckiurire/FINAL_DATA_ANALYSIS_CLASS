```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
top_six_countries= df["job_country"].value_counts().head(6).index
df_median_salary=df.pivot_table(values="salary_year_avg",index="job_country",columns="job_title_short",aggfunc="median")
df_median_salary = df_median_salary.loc[top_six_countries]
job_titles = ["Data Analyst","Data Engineer","Data Scientist"]
df_median_salary=df_median_salary[job_titles]
df_median_salary.plot(kind="bar")
plt.title("Median Annual Salary by Country and Job Title")
plt.xlabel("")
plt.ylabel("Median Annual Salary ($ USD)")
plt.show()
```
